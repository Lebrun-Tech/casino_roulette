```python
#Split a sentence to words
text = input("Enter a sentence: ")
char = input("Enter the character to count: ")
print(f"The character '{char}' is shown: {text.count(char)} {"time" if text.count(char)< 2 else "times"}")
words = text.split()
print(words)
print(f"The sentence has: {len(words)} {"word" if text.count(char)< 2 else "words"}")
```
